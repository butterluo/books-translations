```python
from promptflow.client import load_flow
from promptflow.tracing import start_trace, trace
import os

flow_path = "./translation-standard-flow-v01"
f = load_flow(source=flow_path)

md_file_path = '/Users/huqianghui/Downloads/books-translations/book-chunks'

start_trace()
# 获取所有.md文件名，存储在列表中
md_files = [filename for filename in os.listdir(md_file_path) if filename.endswith('.md')]
# 按文件名升序排序
md_files.sort()

for filename in md_files:
    file_path = os.path.join(md_file_path, filename)
     # Check if it's a file and has a .md extension
    if os.path.isfile(file_path) and filename.endswith('.md'):
        with open(file_path, 'r', encoding='utf-8') as file:
            print(f"reading {filename}")
            md_content = file.read()
        f(english_text=md_content, file_name=filename)
        print(f"Processed {filename}")
```

```python
import tiktoken

encoding = tiktoken.get_encoding("o200k_base")

def num_tokens_from_string(string: str) -> int:
    """Returns the number of tokens in a text string."""
    num_tokens = len(encoding.encode(string))
    return num_tokens

# Directory to save the translation chunks
translation_dir = "/Users/huqianghui/Downloads/books-translations/translated-book-chunks/first-version-translated-book-chunks"

# 获取所有.md文件名，存储在列表中
md_files = [filename for filename in os.listdir(translation_dir) if filename.endswith('.md')]
# 按文件名升序排序
md_files.sort()

for filename in md_files:
    file_path = os.path.join(translation_dir, filename)
     # Check if it's a file and has a .md extension
    if os.path.isfile(file_path) and filename.endswith('.md'):
        with open(file_path, 'r', encoding='utf-8') as file:
            md_content = file.read()
            num_tokens_o200k_base = num_tokens_from_string(md_content)
            print(f"file_name:{filename},gpt4-o token:{num_tokens_o200k_base}...")
```
